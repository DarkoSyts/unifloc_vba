# Просмотрщик результатов массовой адаптации

```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import plotly.plotly as py
import plotly.graph_objs as go
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly import tools
init_notebook_mode(connected=True)

from datetime import date
```

# Объединение входных данных и результатов расчета

```python
input_data = pd.read_csv("input_data.csv", index_col = "Unnamed: 0")
input_data.index = pd.to_datetime(input_data.index)
result_data = pd.read_csv("result_data.csv")
result_data.head()
```

```python
#result_data.head()
```

```python
year_start = 2019
month_start = 2
day_start = 3
hour_start = 0

year_end = 2019
month_end = 2
day_end = 11
hour_end = 0

date_from = pd.to_datetime(pd.DataFrame({'year': [year_start, year_end],
                           'month': [month_start, month_end],
                           'day': [day_start, day_end],
                            'hour': [hour_start, hour_end]}))
day_timestamp_start = date_from.iloc[0]
day_timestamp_end = date_from.iloc[1]
input_data_edited = input_data[(input_data.index >= day_timestamp_start) & (input_data.index <= day_timestamp_end)]
result_data.index = input_data_edited.index
all_data = result_data.join(input_data_edited, lsuffix='_left')
all_data.info()
```
